# Extract the specific educational entities.	

```python
import pandas as pd
import re
import numpy as np
import textstat
from stargazer.stargazer import Stargazer
import statsmodels.api as sm
```

```python
data=pd.read_excel('kickstarter_filtered_1121.xlsx')
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
def extract_education_entities(text):
    # Check if the input is a string
    if not isinstance(text, str):
        return []  # Return an empty list for non-string inputs

    doc = nlp(text)
    education_entities = []
    for ent in doc.ents:
        if ent.label_ in ['ORG', 'GPE']:  # ORG for organizations, GPE for places (could be useful for universities)
            education_entities.append(ent.text)
    return education_entities

# Then apply the function as before
data['education_entities'] = data['Education_Context'].apply(extract_education_entities)
output_file_path = 'output_education_data2.xlsx'

# Save the specified DataFrame columns to an Excel file
data.to_excel(output_file_path, index=False)
```

```python
import re

def extract_institution_names(text):
    if not isinstance(text, str):
        return []

    # Regular expression patterns for educational institution names
    patterns = [
        r'\bUniversity\sof\s[A-Z][a-z]*(?:\s[A-Z][a-z]+)*\b',
        r'\b[A-Z][a-z]+(?:\s[A-Z][a-z]+)*\sUniversity\b',
        #r'\bthe\sUniversity\sof\s[A-Z][a-z]*(?:\s[A-Z][a-z]+)*\b',
        r'\b(?:the\s)?[A-Z][a-z]*(?:\s[A-Z](?:[a-z]*|\.\s))*\s(University|College|Institute|Academy|School)(?:\s\([A-Z]+\))?\b'
        r'\b\w*IT\b',
        r'\b\w*U\b',
        r'\bU[A-Z]+\b',

    ]


    # Find matches
    matches = set()
    for pattern in patterns:
        matches.update(re.findall(pattern, text))

    return list(matches)

# Apply the extraction function to the 'education_entities' column
data2['extracted_institutions'] = data2['education_entities'].apply(extract_institution_names)

data2.to_excel('output_education_data3.xlsx', index=False)
```
